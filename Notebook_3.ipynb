```python
%%writefile app.py
import streamlit as st
import plotly.graph_objects as go
import pandas as pd
import numpy as np


ans = 10

# Function to display the main dashboard
def main_dashboard():
    # Dashboard Title
    st.title("Liquidity Forecasting Dashboard")

    # Subtitle and Bar Chart Section
    st.subheader("Liquidity Ratio Over Time")

    # Generate sample data for a bar chart
    dates = pd.date_range(start="2024-01-01", periods=10)
    liquidity_ratio = np.random.normal(70, 10, size=(10,))

    # Create a bar chart using Plotly
    bar_fig = go.Figure()
    bar_fig.add_trace(go.Bar(x=dates, y=liquidity_ratio, name='Liquidity Ratio'))
    bar_fig.update_layout(title='Historical Liquidity Ratio', xaxis_title='Date', yaxis_title='Ratio')
    st.plotly_chart(bar_fig, key='bar_chart')

    # Section with multiple gauges for key metrics
    st.subheader("Key Metrics")

    # Define metric values (you can replace these with your own data)
    metric_values = [20, 50, 90, 75, 85]  # Current Ratio: 20, Quick Ratio: 50

    # Add custom CSS to reduce spacing and margins
    st.markdown(
        """
        <style>
        .metric-container {
            margin: 0; /* Set margin to 0 */
            padding: 0; /* Set padding to 0 */
        }
        .stPlotlyChart {
            margin: 0; /* Reduce margin for plotly charts */
        }
        </style>
        """,
        unsafe_allow_html=True
    )

    # Create the first row with three metrics
    col1, col2, col3 = st.columns(3)

    with col1:
        fig1 = go.Figure(go.Indicator(
            mode="gauge+number",
            value=10,
            title={'text': "Current Ratio"},
            gauge={
                'axis': {'range': [0, 100]},
                'bar': {'color': 'white', 'thickness': 0.1},
                'steps': [
                    {'range': [0, 30], 'color': 'red'},
                    {'range': [30, 60], 'color': 'yellow'},
                    {'range': [60, 100], 'color': 'green'}
                ]
            }
        ))
        st.plotly_chart(fig1, use_container_width=True, height=250)

    with col2:
        fig2 = go.Figure(go.Indicator(
            mode="gauge+number",
            value=10,
            title={'text': "Quick Ratio"},
            gauge={
                'axis': {'range': [0, 100]},
                'bar': {'color': 'white', 'thickness': 0.1},
                'steps': [
                    {'range': [0, 30], 'color': 'red'},
                    {'range': [30, 60], 'color': 'yellow'},
                    {'range': [60, 100], 'color': 'green'}
                ]
            }
        ))
        st.plotly_chart(fig2, use_container_width=True, height=250)

    with col3:
        fig3 = go.Figure(go.Indicator(
            mode="gauge+number",
            value=metric_values[2],
            title={'text': "Cash Ratio"},
            gauge={
                'axis': {'range': [0, 100]},
                'bar': {'color': 'white', 'thickness': 0.1},
                'steps': [
                    {'range': [0, 30], 'color': 'red'},
                    {'range': [30, 60], 'color': 'yellow'},
                    {'range': [60, 100], 'color': 'green'}
                ]
            }
        ))
        st.plotly_chart(fig3, use_container_width=True, height=250)

    # Create the second row with two metrics
    col4, col5 = st.columns(2)

    with col4:
        fig4 = go.Figure(go.Indicator(
            mode="gauge+number",
            value=metric_values[3],
            title={'text': "Debt to Equity Ratio"},
            gauge={
                'axis': {'range': [0, 100]},
                'bar': {'color': 'white', 'thickness': 0.1},
                'steps': [
                    {'range': [0, 30], 'color': 'red'},
                    {'range': [30, 60], 'color': 'yellow'},
                    {'range': [60, 100], 'color': 'green'}
                ]
            }
        ))
        st.plotly_chart(fig4, use_container_width=True, height=250)

    with col5:
        fig5 = go.Figure(go.Indicator(
            mode="gauge+number",
            value=metric_values[4],
            title={'text': "Operating Cash Flow Ratio"},
            gauge={
                'axis': {'range': [0, 100]},
                'bar': {'color': 'white', 'thickness': 0.1},
                'steps': [
                    {'range': [0, 30], 'color': 'red'},
                    {'range': [30, 60], 'color': 'yellow'},
                    {'range': [60, 100], 'color': 'green'}
                ]
            }
        ))
        st.plotly_chart(fig5, use_container_width=True, height=250)

    # Investment Advice Section
    st.subheader("Investment Advice")

    # Define individual pieces of advice
    advice_list = [
        "Maximizing Profits",
        "Maximising Cash Reserves",
        "Minimizing Risk Exposure"
    ]

    # Create transparent boxes for each piece of advice and a button next to each
    for idx, advice in enumerate(advice_list):
        col1, col2 = st.columns([4, 1])  # Adjust the column sizes for layout

        with col1:
            st.markdown(
                f"""
                <div style="border: 2px solid #4CAF50; border-radius: 5px; padding: 10px; background-color: rgba(255, 255, 255, 0);">
                    {advice}
                </div>
                """,
                unsafe_allow_html=True
            )

        with col2:
            if st.button(f"View Details {idx + 1}", key=f"view_details_{idx}"):
                st.session_state.page = f"details_{idx + 1}"  # Set the page state to the corresponding detail page

# Function to display the details pages
def details_page(page_number):
    if page_number == 1:
        st.title(f"Maximizing Profits")

        if(ans > 0 and ans <= 20):
            st.write("**Convertible Debt Issuance**")
            st.write("")

            st.write("**Strategy:** Issuing convertible debt allows the company to raise funds while offering investors the option to convert the debt into equity at a later date, usually at a predetermined price.")
            st.write("**Upside:** High returns can be realized if the company’s stock price rises above the conversion price. It provides immediate liquidity without the need for immediate equity dilution.")
            st.write("**Risks:** If stock prices do not rise, the company must repay the debt with interest, which could strain cash flow. Additionally, the conversion may dilute existing shareholders if it occurs.")
            st.write("**Time Frame:** Typically medium to long-term, as the conversion usually occurs after a specific period or once certain conditions are met.")

            # Section for Strategic Sales / Divestiture
            st.write("**Strategic Sales / Divestiture**")
            st.write("")

            st.write("**Strategy:** This involves selling non-core or underperforming assets to focus on the company’s primary business areas.")
            st.write("**Upside:** By monetizing non-essential assets, companies can improve profitability and operational efficiency, quickly generating cash.")
            st.write("**Risks:** Selling assets may reduce future revenue potential, and if not done correctly, the company may realize less than expected from the sale.")
            st.write("**Time Frame:** Short to medium-term, depending on the asset market conditions and the complexity of the sales process.")

        if(ans > 20 and ans <= 40):

            # Section for Asset Sale / Divestitures
            st.write("**Asset Sale / Divestitures**")
            st.write("")  # Adding a line break

            st.write("**Strategy:** Selling off assets that are not integral to the core business, realizing their market value.")
            st.write("**Upside:** Generates one-time profits that can be reinvested into more profitable ventures.")
            st.write("**Risks:** Potential future revenue loss from sold assets and the challenge of finding buyers willing to pay fair prices.")
            st.write("**Time Frame:** Medium-term, as finding suitable buyers can take time.")

            # Section for Debt Financing
            st.write("**Debt Financing**")
            st.write("")  # Adding a line break

            st.write("**Strategy:** Acquiring loans or credit facilities to leverage capital for investment in high-yield projects.")
            st.write("**Upside:** Increases potential returns on investment through the tax-deductible interest expense, improving net profits.")
            st.write("**Risks:** Increased debt levels can strain cash flows and heighten financial risk during downturns.")
            st.write("**Time Frame:** Short to medium-term, depending on the financing arrangement and deployment of capital.")

        if(ans > 40 and ans <= 50):

            st.write("**Private Placement**")
            st.write("")  # Adding a line break

            st.write("**Strategy:** Selling securities directly to a select group of investors, typically institutional, rather than through a public offering.")
            st.write("**Upside:** Higher returns possible due to tailored investment terms and less regulatory burden.")
            st.write("**Risks:** Limited liquidity in private placements and potential valuation issues if not enough investors are interested.")
            st.write("**Time Frame:** Short to medium-term, depending on investor interest and transaction complexity.")

            # Section for Structured Debt Financing
            st.write("**Structured Debt Financing**")
            st.write("")  # Adding a line break

            st.write("**Strategy:** Packaging different types of debt instruments with varying risk profiles to attract diverse investors.")
            st.write("**Upside:** Potential for higher returns through complex structuring and premium pricing.")
            st.write("**Risks:** Complexity may lead to higher costs or lower investor confidence if not clearly communicated.")
            st.write("**Time Frame:** Medium-term, as structuring takes time and requires market assessments.")

        if(ans > 50 and ans <= 60):

            # Section for Public Debt Offering
            st.write("**Public Debt Offering**")
            st.write("")  # Adding a line break

            st.write("**Strategy:** Issuing bonds to the public market, allowing access to larger pools of capital.")
            st.write("**Upside:** Interest rate management can enhance profits, while increasing credibility in the market.")
            st.write("**Risks:** Market volatility can affect bond pricing, and reliance on fixed-income may reduce flexibility.")
            st.write("**Time Frame:** Medium-term, usually within a few months for the issuance process.")

            # Section for Secondary Equity Offering (FPO)
            st.write("**Secondary Equity Offering (FPO)**")
            st.write("")  # Adding a line break

            st.write("**Strategy:** Issuing additional shares in the market to raise capital while capitalizing on favorable market conditions.")
            st.write("**Upside:** Directly boosts cash reserves and provides funds for growth or debt repayment.")
            st.write("**Risks:** Potential dilution of existing shares can deter some investors, and success depends on market sentiment.")
            st.write("**Time Frame:** Short to medium-term, as issuance typically occurs within a few weeks to months.")

        if(ans > 60 and ans <= 70):
            st.write("**Secondary Equity Offering (FPO)**")
            st.write("")  # Adding a line break

            st.write("**Strategy:** Leveraging market conditions to issue new shares for capital.")
            st.write("**Upside:** Quick access to funds can enable strategic investments or acquisitions.")
            st.write("**Risks:** Potential negative market perception from existing shareholders, leading to volatility.")
            st.write("**Time Frame:** Short to medium-term, with swift execution possible depending on market sentiment.")

            # Section for Structured Equity Financing
            st.write("**Structured Equity Financing**")
            st.write("")  # Adding a line break

            st.write("**Strategy:** Using equity derivatives or other structured financial products to raise capital.")
            st.write("**Upside:** Enhanced flexibility and potential for increased capital through tailored financing solutions.")
            st.write("**Risks:** Complexity in terms could lead to misunderstandings among investors, potentially affecting demand.")
            st.write("**Time Frame:** Medium-term, as structuring and negotiations take time.")


        if(ans > 70 and ans <= 80):
            st.write("**Structured Equity Financing**")
            st.write("")  # Adding a line break

            st.write("**Strategy:** Utilizing complex financial instruments to maximize capital raising while balancing risk.")
            st.write("**Upside:** Enhanced returns through tailored financing can attract sophisticated investors.")
            st.write("**Risks:** Potential misalignment of interests and confusion regarding terms can arise.")
            st.write("**Time Frame:** Medium-term, requiring substantial structuring time.")

            # Section for Strategic Mergers/Acquisitions
            st.write("**Strategic Mergers/Acquisitions**")
            st.write("")  # Adding a line break

            st.write("**Strategy:** Acquiring or merging with complementary businesses to boost profitability.")
            st.write("**Upside:** Increased market share, synergies, and cost efficiencies can lead to substantial profit growth.")
            st.write("**Risks:** Cultural integration challenges and potential overvaluation of target companies.")
            st.write("**Time Frame:** Long-term, as mergers and acquisitions typically involve extensive due diligence and integration efforts.")


        if(ans > 80 and ans <= 90):
            st.write("**Strategic Mergers/Acquisitions**")
            st.write("")  # Adding a line break

            st.write("**Strategy:** Acquiring firms that complement existing business operations to drive profitability.")
            st.write("**Upside:** Enhanced market position and revenue potential through synergies.")
            st.write("**Risks:** Potential miscalculations in the acquisition process leading to losses.")
            st.write("**Time Frame:** Long-term, with value realization often requiring extensive integration efforts.")

            # Section for Private Equity Investment
            st.write("**Private Equity Investment**")
            st.write("")  # Adding a line break

            st.write("**Strategy:** Engaging with private equity firms to inject capital in exchange for a stake in the business.")
            st.write("**Upside:** Access to significant funds without immediate repayment obligations.")
            st.write("**Risks:** Loss of control and direction, depending on the terms of the investment.")
            st.write("**Time Frame:** Long-term.")


        if(ans > 90 and ans <= 100):
              st.write("**Strategic Mergers/Acquisitions**")
              st.write("")  # Adding a line break

              st.write("**Strategy:** Acquiring firms that complement existing business operations to drive profitability.")
              st.write("**Upside:** Enhanced market position and revenue potential through synergies.")
              st.write("**Risks:** Potential miscalculations in the acquisition process leading to losses.")
              st.write("**Time Frame:** Long-term, with value realization often requiring extensive integration efforts.")

              # Section for Private Equity Investment
              st.write("**Private Equity Investment**")
              st.write("")  # Adding a line break

              st.write("**Strategy:** Engaging with private equity firms to inject capital in exchange for a stake in the business.")
              st.write("**Upside:** Access to significant funds without immediate repayment obligations.")
              st.write("**Risks:** Loss of control and direction, depending on the terms of the investment.")
              st.write("**Time Frame:** Long-term.")




    elif page_number == 2:
        st.title(f"Maximising Cash Reserves")

        if ans > 0 and ans <= 20:
          st.write("*Convertible Debt Issuance*")
          st.write("*Strategy:* As previously mentioned, issuing convertible debt provides immediate capital without requiring equity dilution.")
          st.write("*Upside:* Ensures cash inflow for operational needs while maintaining flexibility in financing.")
          st.write("*Risks:* Future conversion could lead to equity dilution, which may affect current shareholders' value if stock performance is poor.")
          st.write("*Time Frame:* Medium to long-term based on conversion timelines.")

          st.write("*Strategic Sales / Divestiture*")
          st.write("*Strategy:* Selling off non-core assets or divisions that do not align with the company’s long-term goals.")
          st.write("*Upside:* Quick cash inflow helps bolster liquidity and improves the overall balance sheet.")
          st.write("*Risks:* Potential loss of revenue streams from sold assets and the risk of not achieving expected sale prices.")
          st.write("*Time Frame:* Short-term, typically completed within a few months to a year.")



        if(ans > 20 and ans <= 40):
          st.write("*Debt Financing*")
          st.write("*Strategy:* Utilizing loans to enhance liquidity without diluting equity.")
          st.write("*Upside:* Immediate access to capital for operations, expansion, or emergencies without impacting current shareholder stakes.")
          st.write("*Risks:* Increased debt obligations can create long-term cash flow pressures if revenue does not grow accordingly.")
          st.write("*Time Frame:* Immediate to short-term, as funds are usually available quickly.")

          st.write("*Asset Sale / Divestitures*")
          st.write("*Strategy:* Selling off non-essential business units or physical assets for cash.")
          st.write("*Upside:* Quick infusion of cash strengthens liquidity and can help address immediate financial obligations.")
          st.write("*Risks:* Risk of selling assets below market value in a hurry and potential operational disruption.")
          st.write("*Time Frame:* Short-term, typically completed in a few months.")

        if ans > 40 and ans <= 50:
          st.write("*Structured Debt Financing*")
          st.write("*Strategy:* Offering debt instruments with varied risk and return profiles to increase capital inflow.")
          st.write("*Upside:* Flexible repayment options enhance cash reserves while providing immediate liquidity.")
          st.write("*Risks:* Complexity in repayment structures could lead to challenges in cash management.")
          st.write("*Time Frame:* Medium-term, typically aligned with the repayment schedule.")

          st.write("*Public Debt Offering*")
          st.write("*Strategy:* Issuing bonds or other debt instruments to the public market to raise funds.")
          st.write("*Upside:* Large capital inflow without equity dilution and favorable terms aligned with market conditions.")
          st.write("*Risks:* Public offerings can be affected by market conditions, and interest rates may increase during the issuance period.")
          st.write("*Time Frame:* Medium-term, as the process can take several months to complete.")

        if(ans > 50 and ans <= 60):
          st.write("*Secondary Equity Offering (FPO)*")
          st.write("*Strategy:* Raising additional equity capital through a public offering of existing shares.")
          st.write("*Upside:* Provides immediate cash for operational needs, capital projects, or debt reduction.")
          st.write("*Risks:* Existing shareholders may experience dilution in their ownership percentages.")
          st.write("*Time Frame:* Short-term, with the process often completed quickly depending on market conditions.")

          st.write("*Structured Debt Financing*")
          st.write("*Strategy:* Offering various debt securities tailored to meet investor preferences for risk and return.")
          st.write("*Upside:* Attracts diverse investor interest, enhancing liquidity without equity dilution.")
          st.write("*Risks:* Complexity in structuring could complicate repayment and cash management.")
          st.write("*Time Frame:* Medium-term, as structuring takes time and market assessments.")

        if(ans > 60 and ans <= 70):
          st.write("*Structured Equity Financing*")
          st.write("*Strategy:* Utilizing innovative equity instruments, such as preferred stocks or equity-linked notes.")
          st.write("*Upside:* Provides capital while minimizing immediate cash outflows and maintaining flexibility.")
          st.write("*Risks:* Complexity in structure could lead to mispricing and misalignment with investor expectations.")
          st.write("*Time Frame:* Medium-term, typically requiring negotiation and structuring time.")

          st.write("*Debt Restructuring*")
          st.write("*Strategy:* Refinancing existing debt to improve terms or consolidate obligations.")
          st.write("*Upside:* Can lower interest expenses and extend repayment terms, enhancing liquidity.")
          st.write("*Risks:* Potentially adverse effects on credit ratings and future borrowing costs if not managed well.")
          st.write("*Time Frame:* Medium-term, often taking several months to complete negotiations and paperwork.")

        if(ans > 70 and ans <= 80):
          st.write("*Strategic Mergers/Acquisitions*")
          st.write("*Strategy:* Consolidating resources with other firms to enhance cash flow and operational efficiency.")
          st.write("*Upside:* Shared resources can lead to lower operational costs and increased cash reserves.")
          st.write("*Risks:* Integration challenges may lead to initial cash flow strains, affecting short-term liquidity.")
          st.write("*Time Frame:* Long-term, often requiring several quarters to realize cash flow benefits.")

          st.write("*Capital Raising through Public Markets*")
          st.write("*Strategy:* Tapping into public equity markets for raising funds.")
          st.write("*Upside:* Access to vast investor pools can provide substantial cash inflow.")
          st.write("*Risks:* Market conditions can affect capital-raising success and share price volatility.")
          st.write("*Time Frame:* Short to medium-term, depending on market conditions.")

        if(ans > 80 and ans <= 100):
          st.write("*Private Equity Investment*")
          st.write("*Strategy:* Partnering with private equity firms to raise funds for strategic initiatives.")
          st.write("*Upside:* Immediate access to capital can bolster cash reserves for expansion or debt reduction.")
          st.write("*Risks:* Potential loss of operational autonomy based on private equity influence.")
          st.write("*Time Frame:* Long-term, as the partnership typically lasts for several years.")

          st.write("*Strategic Mergers/Acquisitions*")
          st.write("*Strategy:* Utilizing mergers and acquisitions to improve operational cash flow and efficiencies.")
          st.write("*Upside:* Increased scale and scope can lead to greater cash generation capabilities.")
          st.write("*Risks:* Integration challenges may result in temporary cash flow disruptions.")
          st.write("*Time Frame:* Long-term, with benefits realized post-integration.")

    elif page_number == 3:
        st.title(f"Minimizing Risk Exposure")

        if(ans > 0 and ans <= 20):
            st.write("Convertible Debt Issuance")
            st.write("")

            st.write("Strategy: Convertible debt reduces reliance on equity financing while offering investors a safety net through conversion options.")
            st.write("Upside: Provides a buffer against significant capital market fluctuations while retaining the ability to raise funds.")
            st.write("Risks: If the company performs poorly, it may face challenges in servicing the debt without causing liquidity issues.")
            st.write("Time Frame: Medium to long-term, similar to profitability aspects.")

            # Strategic Sales / Divestiture
            st.write("Strategic Sales / Divestiture")
            st.write("")

            st.write("Strategy: Reducing exposure to volatile sectors by selling non-core businesses or assets.")
            st.write("Upside: Minimizes risk by focusing on stable revenue-generating units and shedding those with unpredictable cash flows.")
            st.write("Risks: The company may lose potential upside from future growth in sold areas and face transaction costs.")
            st.write("Time Frame: Short to medium-term, depending on market conditions for asset sales.")

        if(ans >20 and ans <= 40):

            st.write("Private Placement")
            st.write("")

            st.write("Strategy: Raising capital through a private placement to institutional investors, reducing public market volatility risks.")
            st.write("Upside: Lower exposure to regulatory scrutiny and market fluctuations due to selective targeting.")
            st.write("Risks: The lack of liquidity in private investments could pose risks if quick capital access is needed.")
            st.write(":red[Risk Highlight: Potential liquidity risks if quick access to capital is needed.]")  # Highlighting the risk
            st.write("Time Frame: Short to medium-term, depending on investor engagement.")

            # Public Debt Offering
            st.write("Public Debt Offering")
            st.write("")

            st.write("Strategy: Issuing bonds to the public market to secure funding while maintaining fixed-income security exposure.")
            st.write("Upside: Enhances credit ratings with successful issuances, providing long-term financial stability.")
            st.write("Risks: Potential market fluctuations could affect bond prices, impacting future financing costs.")
            st.write("Time Frame: Medium-term, influenced by market conditions during issuance.")

        if(ans > 40 and ans <= 50):

          st.write("### Private Placement")
          st.write("")
          st.write("Strategy: Raising capital through a private placement to institutional investors, reducing public market volatility risks.")
          st.write("Upside: Lower exposure to regulatory scrutiny and market fluctuations due to selective targeting.")
          st.write(":green[Upside Highlight: Lower exposure to regulatory scrutiny and market fluctuations.]")  # Highlighting the upside
          st.write("Risks: The lack of liquidity in private investments could pose risks if quick capital access is needed.")
          st.write(":red[Risk Highlight: Lack of liquidity may pose challenges in accessing quick capital.]")  # Highlighting the risk
          st.write("Time Frame: Short to medium-term, depending on investor engagement.")

          # Public Debt Offering
          st.write("### Public Debt Offering")
          st.write("")
          st.write("Strategy: Issuing bonds to the public market to secure funding while maintaining fixed-income security exposure.")
          st.write("Upside: Enhances credit ratings with successful issuances, providing long-term financial stability.")
          st.write(":green[Upside Highlight: Enhances credit ratings, contributing to long-term stability.]")  # Highlighting the upside
          st.write("Risks: Potential market fluctuations could affect bond prices, impacting future financing costs.")
          st.write(":red[Risk Highlight: Market fluctuations may impact bond prices and future costs.]")  # Highlighting the risk
          st.write("Time Frame: Medium-term, influenced by market conditions during issuance.")


        if(ans > 50 and ans <= 60):

            st.write("### Public Debt Offering")
            st.write("")
            st.write("Strategy: Issuing bonds to diversify funding sources and minimize reliance on equity.")
            st.write("Upside: Reduces equity dilution risk while providing fixed income for financial management.")
            st.write(":green[Upside Highlight: Minimizes equity dilution and secures predictable fixed income.]")  # Highlight the upside
            st.write("Risks: Long-term liabilities can accumulate if revenues do not meet projections.")
            st.write(":red[Risk Highlight: Potential risk of accumulating long-term liabilities.]")  # Highlight the risk
            st.write("Time Frame: Medium-term, typically aligned with the issuance schedule.")

            # Structured Debt Financing
            st.write("### Structured Debt Financing")
            st.write("")
            st.write("Strategy: Creating structured financial instruments to manage risk while raising capital.")
            st.write("Upside: Tailored risk profiles can attract a wider range of investors.")
            st.write(":green[Upside Highlight: Customizable risk profiles appeal to a broader investor base.]")  # Highlight the upside
            st.write("Risks: Complexity in repayment structures could create confusion or financial strain.")
            st.write(":red[Risk Highlight: Potential confusion and financial strain due to complex repayment structures.]")  # Highlight the risk
            st.write("Time Frame: Medium-term, as structuring takes time to implement.")

        if(ans > 60 and ans <= 70):

            st.write("### Debt Restructuring")
            st.write("")
            st.write("Strategy: Adjusting existing debt structures to reduce financial burden and improve cash flow.")
            st.write("Upside: Lower costs and improved terms can stabilize finances during downturns.")
            st.write(":green[Upside Highlight: Stabilizes finances with improved terms during downturns.]")  # Highlight the upside
            st.write("Risks: May be viewed negatively by markets, affecting investor confidence.")
            st.write(":red[Risk Highlight: Could negatively impact investor confidence due to market perception.]")  # Highlight the risk
            st.write("Time Frame: Medium to long-term, depending on the complexity of the restructuring process.")

            # Structured Equity Financing
            st.write("### Structured Equity Financing")
            st.write("")
            st.write("Strategy: Issuing convertible preferred shares or equity-linked notes to minimize risk exposure.")
            st.write("Upside: Offers flexibility for future conversions while raising capital.")
            st.write(":green[Upside Highlight: Provides capital flexibility with future conversion options.]")  # Highlight the upside
            st.write("Risks: Potential dilution upon conversion and complexity in terms can deter some investors.")
            st.write(":red[Risk Highlight: Risk of dilution and complex terms may deter certain investors.]")  # Highlight the risk
            st.write("Time Frame: Medium-term, aligned with the issuance schedule.")

        if(ans > 70 and ans <= 80):

            st.write("### Strategic Mergers/Acquisitions")
            st.write("")
            st.write("Strategy: Merging with or acquiring stable companies to diversify risk.")
            st.write("Upside: Reduces dependence on a single revenue source, spreading risk across multiple business lines.")
            st.write(":green[Upside Highlight: Diversifies risk across multiple revenue streams.]")  # Highlight the upside
            st.write("Risks: High acquisition costs and potential culture clashes can offset perceived benefits.")
            st.write(":red[Risk Highlight: Acquisition costs and cultural differences may reduce effectiveness.]")  # Highlight the risk
            st.write("Time Frame: Long-term, with benefits often realized over several years.")

            # Debt Restructuring
            st.write("### Debt Restructuring")
            st.write("")
            st.write("Strategy: Adjusting existing debt agreements to lessen financial burdens.")
            st.write("Upside: Improved cash flow management and reduced interest expenses bolster financial stability.")
            st.write(":green[Upside Highlight: Enhances financial stability by improving cash flow and reducing interest expenses.]")  # Highlight the upside
            st.write("Risks: Long-term obligations may become unsustainable if revenues decline.")
            st.write(":red[Risk Highlight: Declining revenues could make long-term obligations unsustainable.]")  # Highlight the risk
            st.write("Time Frame: Medium to long-term, as negotiations can take time.")

        if(ans > 80 and ans <= 90):

            st.write("### Private Equity Investment")
            st.write("")
            st.write("Strategy: Aligning with private equity firms to reduce risk exposure while accessing growth capital.")
            st.write("Upside: Professional management and expertise can lead to better decision-making.")
            st.write(":green[Upside Highlight: Private equity firms provide growth capital and professional expertise, improving decision-making.]")  # Highlight the upside
            st.write("Risks: Dependence on external partners for strategic direction can introduce uncertainty.")
            st.write(":red[Risk Highlight: Relying on external partners for strategic direction may introduce uncertainty.]")  # Highlight the risk
            st.write("Time Frame: Long-term, as investments typically span several years.")

            # Strategic Mergers/Acquisitions
            st.write("### Strategic Mergers/Acquisitions")
            st.write("")
            st.write("Strategy: Merging with stable companies to spread risk across multiple revenue streams.")
            st.write("Upside: Diversification reduces vulnerability to market fluctuations in any single sector.")
            st.write(":green[Upside Highlight: Diversification spreads risk across multiple sectors, reducing exposure to market fluctuations.]")  # Highlight the upside
            st.write("Risks: Cultural misalignment and integration complexities could hinder success.")
            st.write(":red[Risk Highlight: Cultural misalignment and integration challenges could reduce the effectiveness of the merger.]")  # Highlight the risk
            st.write("Time Frame: Long-term, requiring careful planning and execution.")

        if(ans > 90 and ans <= 100):

            st.write("### Private Equity Investment")
            st.write("")
            st.write("Strategy: Aligning with private equity firms to reduce risk exposure while accessing growth capital.")
            st.write("Upside: Professional management and expertise can lead to better decision-making.")
            st.write(":green[Upside Highlight: Private equity firms provide growth capital and professional expertise, improving decision-making.]")  # Highlight the upside
            st.write("Risks: Dependence on external partners for strategic direction can introduce uncertainty.")
            st.write(":red[Risk Highlight: Relying on external partners for strategic direction may introduce uncertainty.]")  # Highlight the risk
            st.write("Time Frame: Long-term, as investments typically span several years.")

            # Strategic Mergers/Acquisitions
            st.write("### Strategic Mergers/Acquisitions")
            st.write("")
            st.write("Strategy: Merging with stable companies to spread risk across multiple revenue streams.")
            st.write("Upside: Diversification reduces vulnerability to market fluctuations in any single sector.")
            st.write(":green[Upside Highlight: Diversification spreads risk across multiple sectors, reducing exposure to market fluctuations.]")  # Highlight the upside
            st.write("Risks: Cultural misalignment and integration complexities could hinder success.")
            st.write(":red[Risk Highlight: Cultural misalignment and integration challenges could reduce the effectiveness of the merger.]")  # Highlight the risk
            st.write("Time Frame: Long-term, requiring careful planning and execution.")

    # Button to go back to the main dashboard
    if st.button("Back to Dashboard"):
        st.session_state.page = "main"  # Set the page state to main

# Main logic to check the current page state
if "page" not in st.session_state:
    st.session_state.page = "main"  # Default page

if st.session_state.page == "main":
    main_dashboard()
else:
    page_number = int(st.session_state.page.split("_")[-1])  # Get the page number from the state
    details_page(page_number)
```
